```python
data = {'order_id': [1, 2, 3, 4, 5, 6],
        'customer_number': [101, 102, 101, 103, 102, 101]}
Orders = pd.DataFrame(data)
Orders
```

```python
data = {
    'name': ['CountryA', 'CountryB', 'CountryC', 'CountryD'],
    'population': [50000000, 1000000, 1500000, 30000000],
    'area': [5000000, 2000000, 2900000, 4000000]
}
Countries = pd.DataFrame(data)
Countries
```

```python
big_countries = Countries[(Countries['area'] >= 3000000) | (Countries['population'] >= 25000000)]

print(big_countries)
```

```python
data = {
    'player_id': [1, 1, 2, 3, 3],
    'device_id': [2, 2, 3, 1, 4],
    'event_date': ['2016-03-01', '2016-05-02', '2017-06-25', '2016-03-02', '2018-07-03'],
    'games_played': [5, 6, 1, 0, 5]
}
Activity = pd.DataFrame(data)

Activity
```

```python
first_logins = Activity.groupby('player_id')['event_date'].min().reset_index()
first_logins.columns = ['player_id', 'first_login']

print(first_logins)
```

```python
product_data = {
    'product_id': [1, 2, 3],
    'product_name': ['S8', 'G4', 'iPhone'],
    'unit_price': [1000, 800, 1400]
}
sales_data = {
    'seller_id': [1, 1, 2, 3],
    'product_id': [1, 2, 2, 3],
    'buyer_id': [1, 2, 3, 4],
    'sale_date': ['2019-01-21', '2019-02-17', '2019-06-02', '2019-05-13'],
    'quantity': [2, 1, 1, 2],
    'price': [2000, 800, 800, 2800]
}
Product = pd.DataFrame(product_data)
Sales = pd.DataFrame(sales_data)

Product
Sales
```

```python
merged = Product.merge(Sales,left_on = 'product_id', right_on = 'product_id')
merged
```
